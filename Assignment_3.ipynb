```python
## Question 25) Draw a Pie Chart for the following data
## A)
import matplotlib.pyplot as plt

workers = [20,36,16,5,49]
Section = ['A','B','C','D','E']

plt.pie(workers,labels=Section)
plt.legend(Section)
plt.title('Total workers per section')
plt.show()


## B)
Subject = ['Chemistry','Electronics','Mathematics','Physics','statistics']
Students = [23.17,13.51,17.37,18.92,27.03]    ## Students are in Precentage
plt.pie(Students, labels=Subject)
plt.legend(Subject)
plt.title('Total students per class')
plt.show()
```

```python

## Question 26) Draw a subdivided bar diagram for the data given below: 
import matplotlib.pyplot as plt

Year = ['2017-18','2018-19','2019-20']
Arts = [750,725,850]
Commerce = [1450,1625,1750]

plt.bar(Year,Arts,0.4,label="Arts")
plt.bar(Year,Commerce,0.4,bottom=Arts,label="Commerce")

plt.xlabel("Years")
plt.ylabel("No. of students")
plt.legend(['Arts','Commerce'])
plt.title("Years vs No. of students" )
 
plt.show()
```

```python

## Question 27)Draw a multiple bar diagram for the data given below:


import matplotlib.pyplot as plt
import numpy as np
w=0.4
Years = ['2016','2017','2018','2019','2020']
Turnover = [6.5,6.7,6.9,7,8]   ## In lakhs
Profit = [1.2,1.7,1.8,1.9,2]   ## In lakhs

turnover_bar=np.arange(len(Years))
profit_bar=[i+w for i in turnover_bar]
print(turnover_bar)
plt.bar(turnover_bar,Turnover,w,label="Turnover")
plt.bar(profit_bar,Profit,w,label="Profit")

plt.xlabel("Years")
plt.ylabel("Turnover and profit (in lakhs)")
 
plt.title("Business")
plt.legend()
plt.show()
```

```python
##Q31 ] Generate 60 random numbers based on X random variable, where X follows U(0,1)distribution and calculates Y=-0.8X. Using this simulated data, draw a scatter plot and interpret it.
import numpy as np
x=np.random.uniform(0,1,size=60)
y=-0.8*x

plt.scatter(x,y)
```
